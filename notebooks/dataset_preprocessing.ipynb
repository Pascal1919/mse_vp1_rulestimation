```python
import pandas as pd
import numpy as np
import os
import sys

project_path = r'C:/Users/palig/MSE/Vertiefungsprojekt1/mse_vp1_rulestimation'

# Add the path to sys.path if it does not already exist
if project_path not in sys.path:
    sys.path.append(project_path)

data_path = os.path.join(project_path, 'data')

df_train = pd.read_csv(os.path.join(data_path, 'kaggle', 'train_data', 'train_data.csv'))
df_test = pd.read_csv(os.path.join(data_path, 'kaggle', 'test_data', 'test_data.csv'))

#Drop empty columns
df_train = df_train.drop(columns=['sensor_24', 'sensor_25', 'sensor_26', 'sensor_27', 'RUL'])
df_test =df_test.drop(columns=['sensor_24', 'sensor_25', 'sensor_26', 'sensor_27'])
columns = df_train.columns

# Original CMAPSS data sets for comparison of statistical values
df1 = pd.read_csv(os.path.join(data_path, 'CMAPSS', 'train_FD001.txt'), delimiter=' ')
df2 = pd.read_csv(os.path.join(data_path, 'CMAPSS', 'train_FD002.txt'), delimiter=' ')
df3 = pd.read_csv(os.path.join(data_path, 'CMAPSS', 'train_FD003.txt'), delimiter=' ')
df4 = pd.read_csv(os.path.join(data_path, 'CMAPSS', 'train_FD004.txt'), delimiter=' ')

df1.columns = columns
df2.columns = columns
df3.columns = columns
df4.columns = columns

def drop_empty(df):
    return df.drop(columns=['sensor_22', 'sensor_23'])

df_train = drop_empty(df_train)
df1 = drop_empty(df1)
df2 = drop_empty(df2)
df3 = drop_empty(df3)
df4 = drop_empty(df4)
df_test = drop_empty(df_test)
```

```python
import pandas as pd

def compare_statistics(*dataframes):

    means = {}
    stds = {}

    for i, df in enumerate(dataframes, start=1):
        desc = df.describe()
        means[f'df{i}_mean'] = desc.loc['mean']
        stds[f'df{i}_std'] = desc.loc['std']

    comparison_means = pd.DataFrame(means)
    comparison_stds = pd.DataFrame(stds)

    return comparison_means, comparison_stds

# Compare the statistics
comparison_means, comparison_stds = compare_statistics(df1, df2, df3, df4, df_train, df_test)

# Display the results
print("Comparison of mean values:")
display(comparison_means)

print("\nComparison of standard deviations:")
display(comparison_stds)
```

```python
import pandas as pd
import numpy as np

# Function for calculating the mean values
def calculate_means(df1, df2):
    combined_data = pd.concat([df1, df2], ignore_index=True)
    feature_columns = [col for col in combined_data.columns if 'sensor' in col or 'op_setting' in col]
    means = combined_data[feature_columns].mean()
    return means.to_dict()

# Calculate mean values for the two groups
mean_group1 = calculate_means(df1, df3)
mean_group2 = calculate_means(df2, df4)

# List of relevant columns


def split_data(df, mean_group1, mean_group2):

    feature_columns = [col for col in df.columns if 'sensor' in col or 'op_setting' in col]
    # Group by `engine_no` and calculate mean values per group
    engine_means = df.groupby('engine_no')[feature_columns].mean()

    # Average values as arrays for distance calculation
    mean_values1 = np.array([mean_group1[sensor] for sensor in feature_columns])
    mean_values2 = np.array([mean_group2[sensor] for sensor in feature_columns])

    tmp = pd.DataFrame()
    # Calculation of distances
    tmp['distance_to_group1'] = np.linalg.norm(engine_means.to_numpy() - mean_values1, axis=1)
    tmp['distance_to_group2'] = np.linalg.norm(engine_means.to_numpy() - mean_values2, axis=1)

    engine_means['distance_to_group1'] = tmp['distance_to_group1']
    engine_means['distance_to_group2'] = tmp['distance_to_group2']

    # Group assignment based on shorter distance
    engine_means['group'] = np.where(
        engine_means['distance_to_group1'] < engine_means['distance_to_group2'], 
        'group1', 
        'group2'
    )

    # Check whether the group assignment has worked
    print(engine_means[['distance_to_group1', 'distance_to_group2', 'group']].head())

    # Add the group assignment to `df`.
    df = df.merge(engine_means[['group']], on='engine_no', how='left')

    # Split the DataFrame according to groups
    if 'group' in df.columns:
        group1_data = df[df['group'] == 'group1']
        group2_data = df[df['group'] == 'group2']
    else:
        raise ValueError("Column 'group' was not created correctly.")

    # Output of the group information
    print(f"Size of Group 1: {group1_data.shape}")
    print(f"Size of Group 2: {group2_data.shape}")

    group1_data = group1_data.drop('group', axis=1)
    group2_data = group2_data.drop('group', axis=1)

    id_group1 = group1_data['engine_no'].unique()
    id_group2 = group2_data['engine_no'].unique()
    

    return group1_data, group2_data, id_group1, id_group2

df_train_13, df_train_24, id_train_13, id_train_24 = split_data(df_train, mean_group1, mean_group2)
df_test_13, df_test_24, id_test_13, id_test_24 = split_data(df_test, mean_group1, mean_group2)

display(df_train_13)
```

```python
def normalisation(data_train, data_test):
    # engineID and working time cycle, no need to norm
    prefix_train = data_train.iloc[:, [0, 1]].values
    prefix_test = data_test.iloc[:, [0, 1]].values

    # operational settings and raw sensor data
    inputs_train = data_train.iloc[:, 2:].values
    inputs_test = data_test.iloc[:, 2:].values
    # to avoid devide zero
    eps = 1e-12

    combined = np.concatenate([inputs_train, inputs_test], axis=0)
    min_values = np.min(combined, axis=0)
    max_values = np.max(combined, axis=0)

    normed_data = (inputs_train - min_values) / (max_values - min_values + eps)
    normed_test = (inputs_test - min_values) / (max_values - min_values + eps)

    output_train = np.concatenate((prefix_train, normed_data), axis=1)
    output_test = np.concatenate((prefix_test, normed_test), axis=1)

    return output_train, output_test



df_train_13_normalized, df_test_13_normalized = normalisation(df_train_13, df_test_13)
df_train_24_normalized, df_test_24_normalized = normalisation(df_train_24, df_test_24)
```

```python
# Save preprocessed data for training
np.savetxt(os.path.join(data_path, 'final', 'train_13.txt'), df_train_13_normalized, fmt="%.6f", delimiter=" ")
np.savetxt(os.path.join(data_path, 'final', 'train_24.txt'), df_train_24_normalized, fmt="%.6f", delimiter=" ")
np.savetxt(os.path.join(data_path, 'final', 'test_13.txt'), df_test_13_normalized, fmt="%.6f", delimiter=" ")
np.savetxt(os.path.join(data_path, 'final', 'test_24.txt'), df_test_24_normalized, fmt="%.6f", delimiter=" ")
```
